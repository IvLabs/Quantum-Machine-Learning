```
from numpy import pi
from qiskit import QuantumCircuit
from qiskit.circuit import ParameterVector
```

```
# Setting Default Parameters
num_qubits = 4
reps = 1
x = ParameterVector(name='x', length=num_qubits)
```

## Circuit Implementations

```
# Circuit 1

qc1 = QuantumCircuit(num_qubits)

for _ in range(reps):
    for i in range(num_qubits):
        qc1.rx(x[i], i)
        qc1.rz(x[i], i)
    qc1.barrier()

qc1.draw(output='mpl')
```

```
# Circuit 2

qc2 = QuantumCircuit(num_qubits)

for _ in range(reps):
    for i in range(num_qubits):
        qc2.rx(x[i], i)
        qc2.rz(x[i], i)
    qc2.barrier()
    for control in range(num_qubits-1, 0, -1):
        target = control - 1
        qc2.cx(control, target)
    qc2.barrier()

qc2.draw(output='mpl')
```

```
# Circuit 3

qc3 = QuantumCircuit(num_qubits)

for _ in range(reps):
    for i in range(num_qubits):
        qc3.rx(x[i], i)
        qc3.rz(x[i], i)
    qc3.barrier()
    for control in range(num_qubits-1, 0, -1):
        target = control - 1
        qc3.crz(x[target], control, target)
    qc3.barrier()

qc3.draw(output='mpl')
```

```
# Circuit 4

qc4 = QuantumCircuit(num_qubits)

for _ in range(reps):
    for i in range(num_qubits):
        qc4.rx(x[i], i)
        qc4.rz(x[i], i)
    qc4.barrier()
    for control in range(num_qubits-1, 0, -1):
        target = control - 1
        qc4.crx(x[target], control, target)
        qc4.barrier()

qc4.draw(output='mpl')
```

```
#Circuit 5

qc5 = QuantumCircuit(4)

for _ in range(reps):
    for i in range(num_qubits):
        qc5.rx(x[i], i)
        qc5.rz(x[i], i)
    qc5.barrier()
    for control in range(num_qubits-1, -1, -1):
        for target in range(num_qubits-1, -1, -1):
            if target != control:
                qc5.crz(x[target], control, target)
    qc5.barrier()
    for i in range(num_qubits):
        qc5.rx(x[i], i)
        qc5.rz(x[i], i)
    qc5.barrier()

qc5.draw(output='mpl')
```

```
#Circuit 6

qc6 = QuantumCircuit(4)

for _ in range(reps):
    for i in range(num_qubits):
        qc6.rx(x[i], i)
        qc6.rz(x[i], i)
    qc6.barrier()
    for control in range(num_qubits-1, -1, -1):
        for target in range(num_qubits-1, -1, -1):
            if target != control:
                qc6.crx(x[target], control, target)
                qc6.barrier()
    for i in range(num_qubits):
        qc6.rx(x[i], i)
        qc6.rz(x[i], i)
    qc6.barrier()

qc6.draw(output='mpl')
```

```
#Circuit 7

qc7 = QuantumCircuit(4)

for _ in range(reps):
    for i in range(num_qubits):
        qc7.rx(x[i], i)
        qc7.rz(x[i], i)
    qc7.barrier()
    for control in range(num_qubits-1, 0, -1):
        if control % 2:
            target = control - 1
            qc7.crz(x[target], control, target)
    qc7.barrier()
    for i in range(num_qubits):
        qc7.rx(x[i], i)
        qc7.rz(x[i], i)
    qc7.crz(x[1], 2, 1)
    qc7.barrier()

qc7.draw(output='mpl')
```

```
#Circuit 9

qc9 = QuantumCircuit(4)

for _ in range(reps):
    for i in range(num_qubits):
        qc9.h(i)
    qc9.barrier()
    for control in range(num_qubits-1, 0, -1):
        target = control - 1
        qc9.cz(control, target)
    qc9.barrier()
    for i in range(num_qubits):
        qc9.rx(x[i], i)
    qc9.barrier()

qc9.draw(output='mpl')
```

```
#Circuit 10

qc10 = QuantumCircuit(4)

for _ in range(reps):
    for i in range(num_qubits):
        qc10.ry(x[i], i)
    qc10.barrier()
    for control in range(num_qubits-1, 0, -1):
        target = control - 1
        qc10.cz(control, target)
    qc10.cz(0, num_qubits-1)
    qc10.barrier()
    for i in range(num_qubits):
        qc10.ry(x[i], i)
    qc10.barrier()

qc10.draw(output='mpl')
```

```
#Circuit 13

qc13 = QuantumCircuit(4)

for _ in range(reps):
    for i in range(num_qubits):
        qc13.ry(x[i], i)
    qc13.barrier()
    qc13.crz(x[0], num_qubits-1, 0)
    for control in range(num_qubits-2, -1, -1):
        target = control + 1
        qc13.crz(x[target], control, target)
    qc13.barrier()
    for i in range(num_qubits):
        qc13.ry(x[i], i)
    qc13.barrier()
    qc13.crz(x[num_qubits-2], num_qubits-1, num_qubits-2)
    qc13.crz(x[0], 0, num_qubits-1)
    for i in range(1, 3):
        qc13.crz(x[i-1], i, i-1)
    qc13.barrier()

qc13.draw(output='mpl')
```
